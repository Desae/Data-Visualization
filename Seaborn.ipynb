### Removing Axes Spines
despine() removes top and right spines by default.

```python
# white, despine
sns.set_style('white')
plt.figure()
x1 = [10, 20, 5, 40, 8]
x2 = [30, 43, 9, 7, 20]
plt.plot(x1, label='Group A')
plt.plot(x2, label='Group B')
sns.despine()
plt.legend()
plt.show()
```

```python
sns.despine(fig=None, ax=None, left=True, right=False, top=False, bottom=True, offset=None, trim=False)
```
