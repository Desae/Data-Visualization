### Plotting Stock prices of the Big-5

```python
# google data
google_df = pd.read_csv('./datasets/stocks_data/GOOGL_data.csv')
google_df.head()
```

```python
# microsft data
ms_df = pd.read_csv('./datasets/stocks_data/MSFT_data.csv')
ms_df.head()
```

```python
# facebook data
fb_df = pd.read_csv('./datasets/stocks_data/FB_data.csv')
fb_df.head()
```

```python
# amazon data
amazone_df = pd.read_csv('./datasets/stocks_data/AMZN_data.csv')
amazone_df.head()
```

```python
# apple
apple_df = pd.read_csv('./datasets/stocks_data/AAPL_data.csv')
apple_df.head()
```

```python
# create figure
plt.figure(figsize=(15,7), dpi=300)

# making plots
plt.plot('date','close', data=google_df, label='Google')
plt.plot('date','close', data=ms_df, label='Microsoft')
plt.plot('date','close', data=fb_df, label='Facebook')
plt.plot('date','close', data=amazone_df, label='Amazon')
plt.plot('date','close', data=apple_df, label='Apple')

# Specify ticks for x and y axis
plt.xticks(np.arange(0,1260,40), rotation=70)
plt.yticks(np.arange(0,1450,100))

#setting title
plt.title('Stock Trend for the BIG 5', fontsize=16)

# setting ylabel
plt.ylabel('Closing price in $', fontsize=14)

# adding grid
plt.grid()

# adding legend
plt.legend()
```

### Barcharts

```python
# format
# plt.bar(x, height, [width]) - default width is 0.8
# plt.barh(x, height, [width]) - horizontal barchart 
```

```python
plt.bar(['A','B','C','D'],[20,25,40,10])
```

```python
# creating sub-categories
labels = ['A','B','C','D']
x = np.arange(len(labels))
width = 0.4
plt.bar(x-width/2, [20,25,40,10], width=width)
# shifting the second bar by an offset equal to the width of the first bar creates sub-categories
plt.bar(x-width/2 +0.4, [30,15,30,20], width=width)

# Ticks
plt.xticks(x)

ax=plt.gca()
ax.set_xticklabels(labels)
```

### Movie rating activity

```python
movie_df = pd.read_csv('./datasets/movie_scores.csv')
movie_df.head()
```

```python
labels = movie_df['MovieTitle'].to_list()
labels
```

```python
plt.figure(figsize=(10,5))
x = np.arange(len(labels))
width= 0.4
#
plt.bar(x-width/2, movie_df['Tomatometer'] ,width=width, label='Tomatometer')
plt.bar(x-width/2 + 0.4, movie_df['AudienceScore'], width=width, label='AudienceScore')

plt.xticks(x)
ax = plt.gca()
ax.set_xticklabels(labels, rotation=20)

plt.yticks(100)

plt.legend()
plt.grid()
```

```python
# create figure
plt.figure(figsize=(10,5), dpi=300)

# create barplot
pos = np.arange(len(movie_df['MovieTitle']))
width = 0.3

plt.bar(x-width/2, movie_df['Tomatometer'] ,width=width, label='Tomatometer')
plt.bar(x+width/2, movie_df['AudienceScore'], width=width, label='AudienceScore')

# specify ticks
plt.xticks(pos, rotation=10)
plt.yticks(np.arange(0,101,20))

# Get current Axes for setting tick labels and horizontal grid
ax = plt.gca()

# Set tick labels
ax.set_xticklabels(movie_df['MovieTitle'])
ax.set_yticklabels(['0%','20%','40%','60%','80%','100%'])

# Add minor ticks for y-axis in the interval of 5
ax.set_yticks(np.arange(0,100,5), minor=True)

# Add major horizontal grid with solid lines
ax.yaxis.grid(which='major')

# Add minor horizontal grid with dashed lines
ax.yaxis.grid(which='minor', linestyle='--')

# Add title
plt.title('Movie Comparison')

# Add legend
plt.legend()
```
